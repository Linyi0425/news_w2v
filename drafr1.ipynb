```python
def stop_words():
    stop_words_file = open('stop_words_ch.txt', 'r')
    stopwords_list = []
    for line in stop_words_file.readlines():
        stopwords_list.append(line[:-1])
        #print(type(stopwords_list[-1]))
    return stopwords_list

def jieba_fenci(raw, stopwords_list):
    # 使用结巴分词把文件进行切分
    word_list =list(jieba.cut(raw, cut_all=False))
    #print(type(word_list))
    for word in word_list:
        if word in stopwords_list:
            #print('stop')
            word_list.remove(word)
    # word_set用于统计A[nClass]
    #word_list.remove('\n')
    #word_list = ' '.join(word_list)
    #print(type(word_list))
    return word_list
```

```python
stopwords_list = stop_words()

with open('yuliao2.txt','wb') as y:
    for i in range(df.shape[0]):
        w_l = jieba_fenci(df['title'][i],stopwords_list)
        #print(w_l)
        for n in range(len(w_l)):
            y.write(w_l[n].encode('utf-8'))
        y.write('\n'.encode('utf-8'))
        if (i+1)%10000==0:
            #break
            print(i)
        
        
```

```python
import numpy as np
import heapq
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
import time
import jieba
model = Word2Vec.load('train_w2v.model')
vocab = list(model.wv.vocab.keys())
def clear_list(l,vocab):
    
    #print(l)
    l1 = []
    for word in l:
        if word in vocab:
            l1.append(word)
        else:
            l.remove(word)
            #print(word)
    return l1
        
    
        
test = pd.read_csv('test_data.csv',encoding='gbk')
test['top20'] = np.zeros(test.shape[0])
df = pd.read_csv('train_data.csv')
stopwords_list = stop_words()
test_words = []
a = time.time()
for i in range(test.shape[0]):
    raw = test['title'][i]
    l1 = jieba_fenci(raw,stopwords_list)
    l1 = clear_list(l1,vocab)
    test_words.append(l1)
b = time.time()
print(b-a)
anss = np.zeros((50,485686))
print(anss.shape)
for i in range(df.shape[0]):
    
    raw = df['title'][i]
    
    
    l1 = jieba_fenci(raw,stopwords_list)
    
    l1 = clear_list(l1,vocab)
    #print('test',l1)
    for j in range(test.shape[0]):
        l2 = test_words[j]
        try:
            anss[j,i] = model.n_similarity(l1,l2)
        except:anss[j,i] = 0
    if i%1000==0:
        print(i)
        
print('done')
for i in range(50):
    ans = anss[i,:]
    top = heapq.nlargest(20,range(len(ans)),ans.__getitem__)
    test['top20'][i] = str(top)
    
    test.to_csv('top20.csv',mode='a')

    
```
